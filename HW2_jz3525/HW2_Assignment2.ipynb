```python
url = "https://data.cityofnewyork.us/resource/hsry-3x74.json"
```

```python
df = pd.read_json(url)
df.head(7)
```

```python
df = df[['year', 'mean_scale_score','num_level_1']]
```

```python
df.rename(columns  =  {'year':'Year', 
               'mean_scale_score':'Mean Scale Score', 'num_level_1':'Num Level 1'}, inplace=True)
```

```python
df2 = df.plot(x = 'Year', y = 'Mean Scale Score');
df2.set_ylabel("Score")
```

```python
df2 = df.plot(x = 'Year', y = 'Num Level 1');
df2.set_ylabel("Score")
```
